```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0,10,11)
y = x**2
```

```python
x
```

```python
y
```

```python
# functional method of plotting 
plt.plot(x,y)
plt.xlabel('X label')
plt.ylabel('Y label')
plt.title('Title')
```

```python
#multiplot in same frame
plt.subplot(1,2,1)
plt.plot(x,y,'r-')
plt.subplot(1,2,2)
plt.plot(y,x,'b')
```

```python
#object oriented method
fig = plt.figure()
axes = fig.add_axes([0.2,0.2,0.8,0.8])
axes.plot(x,y)
axes.set_xlabel('X label')
axes.set_ylabel('Y label')
axes.set_title('Title')
```

```python
fig = plt.figure()
axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes2 = fig.add_axes([0.2,0.5,0.4,0.3])
axes1.set_xlabel('x label')
axes1.set_ylabel('y label')
axes2.set_xlabel('x label')
axes2.set_ylabel('y label')
axes1.set_title('title')
axes2.set_title('title')
axes1.plot(x,y)
axes2.plot(y,x,'r-')
```

```python
#sub plotting
fig,axes = plt.subplots(nrows=1,ncols=2)

axes[0].plot(x,y)
axes[0].set_title('First plot')

axes[1].plot(y,x,'g')
axes[1].set_title('Second plot')
```

```python
fig,axes = plt.subplots(nrows = 2,ncols=1,figsize=(4,2),dpi=100)
axes[0].plot(x,y)
axes[1].plot(y,x,'r-')
plt.tight_layout()
```

```python
fig.savefig('My_picture.png',dpi=100)
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
#ax.set_title('Title')
ax.plot(x,x**2,label='X squared')
ax.plot(x,x**3,label='X cubed')

ax.legend(loc=0)
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y,color = 'purple',linewidth=3,linestyle='--')
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y,color = 'purple',lw=3,ls='--', marker='o',markersize=10,
        markerfacecolor='red',markeredgewidth=3,markeredgecolor='yellow')
```
